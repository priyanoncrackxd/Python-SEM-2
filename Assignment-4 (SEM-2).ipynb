c) Identify maximum and minimum values in the array

2. Write a program to input a array of numbers from the user (at least 10 elements in list), sort them
and perform slicing operations to get elements between indexes such as 2-5, 5-8, 2-9.

```python
numbers = []
while len(numbers) < 10:
    try:
        num = float(input("Enter a number: "))
        numbers.append(num)
    except ValueError:
        print("Please enter a valid number.")

sorted_numbers = sorted(numbers) 

slices = [(2, 5), (5, 8), (2, 9)]

for start, end in slices:
    sliced_elements = sorted_numbers[start:end]
    print(f"Elements from index {start} to {end - 1}: {sliced_elements}")
```

3. Create an array of random integer numbers as a numpy array, sort them and perform operations
such as reshaping of the array into matrix of feasible dimensions. (e.g., if we have an array of 1 * 10,
then we can reshape it into 2 * 5 or 5 * 2 matrix.) [Hint: Use the array of reshape (row * column) ]

```python
import datetime

def operations(num_list, scalar):
    # function to perform arithmetic operation in the list of numbers
    add_list = []
    sub_list = []
    mul_list = []
    div_list = []
    for n in num_list:
        add_list.append(n + scalar)
        sub_list.append(n - scalar)
        mul_list.append(n * scalar)
        div_list.append(n / scalar)
    return add_list, sub_list, mul_list, div_list

def file_write(num_list, s, result):
    # function to store each operation in a file along with the current date and time
    file = open('operations.txt','a')
    file.write(f"\nDate and Time: {datetime.datetime.now()}\n") # extracts and stores current date and time
    file.write(f"User Input Numbers: {num_list}\n")
    file.write(f"User input scalar: {s}\n")
    file.write(f"Addition: {result[0]}\n")
    file.write(f"Subtraction: {result[1]}\n")
    file.write(f"Multiplication: {result[2]}\n")
    file.write(f"Division: {result[3]}\n")
    file.write(f"---------------------------------------------------------")
    file.close()

def file_read():
    # function to read the data of the file
    f = open('operations.txt','r')
    print(f.read())
    f.close()
             
choice = 'y'
while (choice == 'y'):
    num_input = input("Enter list elements separated by space:")
    n_list = num_input.split()
    num_list = [int(n) for n in n_list] # creates a list of integer with the user input data
    scalar = float(input("Enter a scalar number:"))
    result = operations(num_list, scalar) # function call to perform operations
    file_write(num_list, scalar, result) # function call to write the results in a file
    choice = input("Add another list? [y/n]:")
    
print("\nOperations history:")
file_read() # function call to read all the results from the file 
```

4. Write a program to input 2 matrices of certain dimensions and perform the matrix operations such
as additions, subtraction, multiplication using numpy. Validation of matrix size should be done before
the operations are performed. Mismatch of size for operations should raise the exception.

```python
import numpy as np

def input_matrix(matrix_name):
    rows = int(input(f"Enter the number of rows for {matrix_name}: "))
    cols = int(input(f"Enter the number of columns for {matrix_name}: "))
    matrix = np.array([[int(input(f"Enter element {j+1} for row {i+1}: ")) for j in range(cols)] for i in range(rows)])
    return matrix

def validate_matrices(A, B):
    if A.shape != B.shape:
        raise ValueError("Matrix dimensions do not match")

def matrix_operations():
    A = input_matrix("Matrix A")
    B = input_matrix("Matrix B")
    validate_matrices(A, B)
    print("Matrix A:")
    print(A)
    print("\nMatrix B:")
    print(B)
    print("\nAddition of matrices:")
    print(np.add(A, B))
    print("\nSubtraction of matrices:")
    print(np.subtract(A, B))
    print("\nMultiplication of matrices:")
    print(np.matmul(A, B))

try:
    matrix_operations()
except ValueError as e:
    print(e)
```

5. Write a program to read the csv file “weight_height.csv” using Pandas. Plot the data as a
scatterplot (weight vs height, age vs weight, height vs age, gender vs height, gender vs weight) using
Matplotlib library.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("weight_height.csv") # read the csv file

# extract data
age = df['age']
gender = df['gender']
height = df['height']
weight = df['weight']

fig, axis = plt.subplots(3, 2, figsize=(12, 15)) # preparing for subplots

# Scatter plot of weight vs height
axis[0, 0].scatter(weight, height, color='blue')
axis[0, 0].set_xlabel('Weight')
axis[0, 0].set_ylabel('Height')
axis[0, 0].set_title('Weight vs Height')

# Scatter plot of age vs weight
axis[0, 1].scatter(age, weight, color='green')
axis[0, 1].set_xlabel('Age')
axis[0, 1].set_ylabel('Weight')
axis[0, 1].set_title('Age vs Weight')

# Scatter plot of height vs age
axis[1, 0].scatter(height, age, color='red')
axis[1, 0].set_xlabel('Height')
axis[1, 0].set_ylabel('Age')
axis[1, 0].set_title('Height vs Age')

# Scatter plot of  gender vs height
axis[1, 1].scatter(gender, height, color='purple')
axis[1, 1].set_xlabel('Gender')
axis[1, 1].set_ylabel('Height')
axis[1, 1].set_title('Gender vs Height')

# Scatter plot of gender vs weight
axis[2, 0].scatter(gender, weight, color='orange')
axis[2, 0].set_xlabel('Gender')
axis[2, 0].set_ylabel('Weight')
axis[2, 0].set_title('Gender vs Weight')

# display plot
plt.show()
```

6.Read the data from csv file “weight_height.csv” in a data frame using Pandas. Add 2 additional
columns (BMI and Risk) in the existing dataframe. Add the data according to the calculations given
below.
BMI = Weight / Height
Risk values vary according to the conditions given below:
BMI less than 18.5 : Nutrient deficient
BMI between 18.5 and 24.9: lower risk
BMI between 25 and 29.9: Heart disease risk
BMI between 30 and 34.9: Higher risk of diabetes, heart disease
BMI 40 or higher: Serious health condition risk

```python
import pandas as pd

df = pd.read_csv('weight_height.csv') # reads the csv file

# extracts data
age = df['age']
gender = df['gender']
height = df['height']
weight = df['weight']

# function to categorize risk values based on BMI 
def BMI_category(bmi):
    if (bmi<18.5):
        return 'Nutrition deficient'
    elif (bmi >= 18.5 and bmi <= 24.9):
        return 'lower risk'
    elif (bmi >= 25 and bmi <= 29.9):
        return 'Heart disease risk'
    elif (bmi >= 30 and bmi <= 34.9):
        return 'Higher risk of diabetes, heart disease'
    elif (bmi >= 40):
        return 'Serious health condition risk'

df['BMI'] = weight / ((height/100)) # adds new column named BMI based on calculation
df['risk values'] = df['BMI'].apply(BMI_category) # adds new column 'risk values' by applying function BMI_category to each row

print(df.head(6)) # prints first 6 row data
```
